## Retrieval-Augmented Generation (RAG)

Add knowlegde DB to the LLMs without more training or fine-tuning

Support knowlegde updating

Reduce hallucinations

> Chunking strategy

- Fixed-size - simple, predetermined lenghts

- Recursive, using hierarchical division (mother and son chunks)

- Semantic grouping into larger chuncks

> Metada filtering

- Location/language

- Department

- Categories

- Date

> Retriever

- Dense, using semantic similarities with embedding models (sentence similairy)

- Sparse, using key words (BM25)

- Hybrid, dense + sparse models combined

- Reranking, model reviewing the score after retrieving

> Vector store

**Creation for this application**

- user: saulgoodai

- location: us-east-1

- name: kb-sstedoldi-cv

- IAM permmision role: AmazonBedrockExecutionRoleFor_kb-sstedoldi-cv

- Amazon S3 as data source: s3://sstedoldi-cv as kb-sstedoldi-cv-and-other-files

metadata should be added in the bucket: ".metadata-json"

- parsing: default

- chunking: fixed-size 300 and 10 %

- embedding model: titan text, binary emb (cheaper) and vector dim 256 (less memory)

- vector db: opensearch serverless

**Once created**

- Manual sync - updating the KB

- data sources can be added (max 5 sources), or enlarge adding doc into the buckets

**Testing KB**

Using easy access playground with different options: 

- Test mode: retrieve or retrieve + LLM

- Search type: hydrid or semantic

- Other test: filters, guardrails and reranking (another model to use and to enable)

- Query modification: breaking down queries, for complex questions

**Automatic evalutation**

- an evaluator model is requiered

- an response generator model should be selected

- options to test: retrieval only or retrieval and responses

- Metrics:

    - Context relevance (retrival) - query vs. retrieved context
    - Context coverage (ground truth requiered) - answers/ground truth vs retrieved context

- Quality:

    - Helpfulness
    - Correctness
    - Logical coherence
    - Faithfulness
    - Completeness

- Responsible AI:

    - Harmfulness
    - Refusal
    - Stereotyping
